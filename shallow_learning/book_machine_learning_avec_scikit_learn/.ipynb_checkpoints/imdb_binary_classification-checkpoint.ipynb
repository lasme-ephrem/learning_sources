# Préparation des données

```python
import numpy as np
```

```python
def vectorize_sequences(sequences, dimension = 10000):
    results = np.zeros(shape = (len(sequences), dimension))
    for i, sequence in enumerate(sequences):
        results[i, sequence] = 1.
    return results
```

```python
train_data[:2]
```

```python
vectorize_sequences(train_data[:2])
```

```python
x_train = vectorize_sequences(train_data)
x_test = vectorize_sequences(test_data)
```

```python
np.asarray(train_labels)
```

```python
np.asarray(train_labels).astype("float32")
```

```python
y_train = np.asarray(train_labels).astype("float32")
y_test = np.asarray(test_labels).astype("float32")
```

# Architecture du réseau de neurones

```python
x_train
```

```python
x_train.shape
```

```python
x_train[0]
```

```python
x_train[0].shape #input shape
```

```python
from tensorflow.keras import models
from tensorflow.keras import layers

model = models.Sequential()
model.add(layers.Dense(16, activation = "relu", input_shape = (10000, )))
model.add(layers.Dense(16, activation = "relu"))
model.add(layers.Dense(1, activation = "sigmoid"))
```

```python
model.summary()
```

# Compilation du modèle

```python
model.compile(optimizer = "rmsprop",
             loss = "binary_crossentropy",
             metrics = ["accuracy"])
```

# Approche de validation

```python
x_val = x_train[:10000]
partial_x_train = x_train[10000:]
```

```python
y_val = y_train[:10000]
partial_y_train = y_train[10000:]
```

# Entraînement du modèle

```python
history = model.fit(partial_x_train,
                   partial_y_train,
                   epochs = 20,
                   batch_size = 512,
                   validation_data = (x_val, y_val))
```

# Visualisation de la courbe d'apprentissage

```python
history_dict = history.history
history_dict.keys()
```

```python
import matplotlib.pyplot as plt

loss_values = history_dict["loss"]
val_loss_values = history_dict["val_loss"]
epochs = range(1, len(loss_values) + 1)

plt.figure(dpi = 100)
plt.plot(epochs, loss_values, "bo", label = "Entraînement")
plt.plot(epochs, val_loss_values, "b", label = "Validation")
plt.title("Perte pendant l'entraînement et la validation")
plt.xlabel("Nombre d'époques")
plt.ylabel("Perte")
plt.show()

acc_values = history_dict["accuracy"]
val_acc_values = history_dict["val_accuracy"]

plt.figure(dpi = 100)
plt.plot(epochs, acc_values, "bo", label = "Entraînement")
plt.plot(epochs, val_acc_values, "b", label = "Validation")
plt.title("Exactitude pendant l'entraînement et la validation")
plt.xlabel("Nombre d'époques")
plt.ylabel("Exactitude de prédiction")
plt.show()
```

# Confirmation des résultats et entraînement définitif

```python
model = models.Sequential()
model.add(layers.Dense(16, activation = "relu", input_shape = (10000, )))
model.add(layers.Dense(16, activation = "relu"))
model.add(layers.Dense(1, activation = "sigmoid"))

model.compile(optimizer = "rmsprop",
             loss = "binary_crossentropy",
             metrics = ["accuracy"])

model.fit(partial_x_train,
          partial_y_train,
          epochs = 4,
          batch_size = 512)

results = model.evaluate(x_test, y_test)
```
